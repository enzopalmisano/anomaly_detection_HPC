```python
classifier = load_model("./models/my_models/classifier_model04.h5")
```

```python
nodes = ["r183c09s01","r183c09s03","r183c09s04","r183c10s01","r183c10s02","r183c10s03","r183c11s01","r183c11s02","r183c11s03","r183c11s04","r183c15s01","r183c15s02","r183c15s03","r183c15s04","r183c16s03","r183c09s02","r183c11s04","r183c12s01","r183c12s02","r183c12s03","r183c13s01","r183c13s02","r183c13s03","r183c13s04","r183c14s01","r183c14s02","r183c14s03","r183c14s04"]
```

```python
def assign_class(predicted_probabilities, threshold=0.5):
    classes = []
    for i in range(len(predicted_probabilities)):
        if(predicted_probabilities[i] >= threshold):
            classes.append(1)
        else:
            classes.append(0)
    return np.asarray(classes)
```

```python
import pandas as pd
import glob

path = r'./my_code/test_dataset' # use your path
all_files = glob.glob(path + "/*")
count=0
li = []
table_html="""<table border=1 frame=void rules=all style="width:100%;align="center";"><tr><th>Nodo</th><th>Accuracy</th>
<th>F1Score Normal</th><th>F1Score Anomaly</th><th>F1Score Average</th><th>Recall Normal</th><th>Recall Anomaly</th>
<th>Recall Average</th>
<th>Precision Normal</th><th>Precision Anomaly</th><th>Precision Average</th><th>Confusion Matrix(TN,FN,FP,TP)</th></tr>"""
tot_acc=0
tot_pre=0
tot_recall=0
tot_f1=0
for filename in all_files:
    df = pd.read_csv(filename)
    labels = df["label"]
    df = df.drop("label",axis=1)
    df = df.drop("Unnamed: 0",axis=1)
    predict_proba = classifier.predict(df)
    classes = assign_class(predict_proba)
    table_html+="<tr>"
    table_html+="<td>%s</td>"%nodes[count]
    table_html+="<td>%s</td>"%'%.2f'%(accuracy_score(labels,classes)*100)
    tot_acc+=accuracy_score(labels,classes)*100
    table_html+="<td>%s</td>"%'%.2f'%(f1_score(labels,classes,pos_label=0,average='binary'))
    table_html+="<td>%s</td>"%'%.2f'%(f1_score(labels,classes,pos_label=1,average='binary'))
    table_html+="<td>%s</td>"%'%.2f'%(f1_score(labels,classes,average='weighted'))
    tot_f1+=f1_score(labels,classes,average='weighted')
    table_html+="<td>%s</td>"%'%.2f'%(recall_score(labels,classes,pos_label=0,average='binary'))
    table_html+="<td>%s</td>"%'%.2f'%(recall_score(labels,classes,pos_label=1,average='binary'))
    table_html+="<td>%s</td>"%'%.2f'%(recall_score(labels,classes,average='weighted'))
    tot_recall+=recall_score(labels,classes)
    table_html+="<td>%s</td>"%'%.2f'%(precision_score(labels,classes,pos_label=0,average='binary'))
    table_html+="<td>%s</td>"%'%.2f'%(precision_score(labels,classes,pos_label=1,average='binary'))
    table_html+="<td>%s</td>"%'%.2f'%(precision_score(labels,classes,average='weighted'))
    tot_pre += precision_score(labels,classes)
    table_html+="<td>%s</td>"%confusion_matrix(labels, classes)
    table_html+="</tr>"
    count+=1
table_html+="<tr>"
table_html+="<td>%s</td>"%"AVERAGE: "
table_html+="<td>%s</td>"%'%.2f'%(tot_acc/count)
table_html+="<td>%s</td>"%'%.2f'%(tot_f1/count)
table_html+="<td>%s</td>"%'%.2f'%(tot_recall/count)
table_html+="<td>%s</td>"%'%.2f'%(tot_pre/count)
table_html+="</tr>"
table_html += "</table>"

```
